Import matplotlib

```python
  import matplotlib.pyplot as plt
  import numpy as np
  import pandas as pd
```

Plotting x and y points

```python
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y)
plt.show()
```

Bar Chart

```python
categories=['A','B','C','D','E']
values=[5,2,3,4,6]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='green')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

Histogram

```python
data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]
plt.hist(data, bins=16, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram Example')
plt.show()
```

Scatter plot

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]
plt.scatter(x, y)
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Simple Scatter Plot Example')
plt.show()
```

Pie chart

```python
# Sample data
labels = ['A', 'B', 'C', 'D','E','F','G','H']
sizes = [55, 30, 35, 30,39,34,34,21]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')  # Ensures the pie is drawn as a circle
plt.show()
```

Box plot

```python
data = [7, 8, 4, 6, 2, 9, 2, 5, 7, 8, 10]
plt.boxplot(data)
plt.title('Simple Box Plot Example')
plt.ylabel('Values')
plt.show()
```

Heatmap

```python
data = np.random.rand(2, 10)
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Simple Heatmap Example')
plt.show()
```
